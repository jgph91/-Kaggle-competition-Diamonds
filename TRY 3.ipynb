```python
data = pd.read_csv('./input/data.csv/data.csv')
```

```python
steps = [
    
    StandardScaler(),
    Normalizer(),
    PCA(n_components=2)
]

pipe = make_pipeline(*steps)

```

```python
# [carat] max value abnormally high compared to the mean an 75% quartile, y and z 95%, x y z 5%
stats = data.describe().T
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
data = data[data.carat < np.percentile(data.carat,95)]
```

```python
stats = data.describe().T
stats
```

```python
data = data[data.y < np.percentile(data.y,95)]
data = data[data.z < np.percentile(data.z,95)]
```

```python
data = data[data.x > np.percentile(data.x,5)]
data = data[data.y > np.percentile(data.y,5)]
data = data[data.z > np.percentile(data.z,5)]
```

```python
steps = [
    
    StandardScaler(),
    Normalizer(),
    
]

pipe = make_pipeline(*steps)
X = data[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']]
```

```python
X_dummied = pd.get_dummies(X)
X_dummied
```

```python
#Cleaning the df using functions in steps
pipe = make_pipeline(*steps)
X_feats = pipe.fit_transform(X_dummied)
```

```python
y = data.price
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_feats, y, test_size=0.2)
```

```python
model = LinearSVC()
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test) 


from sklearn.metrics import r2_score
R2 = r2_score(y_test,y_pred)
print(f'R2 = {round(R2,2)}')
```
